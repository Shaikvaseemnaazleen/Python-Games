# COVID - 19 DATA ANALYSIS

## Import the modules required

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

print("Imported the modules required..")
```

## Import the  COVID - 19 data set

```python
# our dataset is a csv file  -> covid19_Confirmed_dataset.csv
# we use read_csv() method to read the contents of the csv file and convert to df

corona_dataset_csv = pd.read_csv("covid19_Confirmed_dataset.csv")

# to view the top 10 lines of dataset -> use head

corona_dataset_csv.head(10)
```

```python
# check the dimensions of dataset -> shape

corona_dataset_csv.shape
```

## Delete the useless columns

```python
# we dont require the longitudes and latitudes for our analysis -> use drop

corona_dataset_csv.drop(['Lat', 'Long'], axis = 1, inplace = True)

# view the changes

corona_dataset_csv.head(10)
```

## Aggregating the rows by country

```python
corona_dataset_aggregated = corona_dataset_csv.groupby("Country/Region").sum()
```

```python
# view the changes made

corona_dataset_aggregated.head(10)
```

```python
# check the shape of our dataframe

corona_dataset_aggregated.shape
```

# calculate and plot the first derivative of the curve

```python
# to see the differnece in infection rate

corona_dataset_aggregated.loc['China'].diff().plot()
```

```python
# to calculate the maximum infection rate in china

corona_dataset_aggregated.loc['China'].diff().max() # the max people affected by corona in one day
```

```python
# now lets calculate for india

corona_dataset_aggregated.loc['India'].diff().max() # the max people affected by corona in one day (india)
```

# Find the maximum infection rate for all countries

```python
# loop through every country and find the max infection rate

max_infection_rate = [] # store the values pf max infection rate
countries = corona_dataset_aggregated.index

for country in countries:
    max_infection_rate.append(corona_dataset_aggregated.loc[country].diff().max())
    
```

```python
# add the column to the dataframe named maximum infection rate

corona_dataset_aggregated['maximum infection rate'] = max_infection_rate
```

```python
# view the added column at the end of dataframe

corona_dataset_aggregated
```

# Create a new dataframe with needed columns

```python
# to create a new dataframe we use DataFrame() method

corona_data = pd.DataFrame(corona_dataset_aggregated['maximum infection rate'])
```

```python
# view the contents of dataframe

corona_data.head(10)
```
